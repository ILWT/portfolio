Также проанализируем время продажи квартиры. После этого исключим из датафрейма редкие и выбивающиеся значения по указанным столбцам.

Мы видим большое количество выбросов в данных. Посчитаем долю значений, расстояние которых от среднего превышает 3 стандартных отклонения (по правилу трех сигм), и в случае, если их количество не превышает 5% от всех значений, отбросим их.

```python
# напишем функцию, которая возвращает по столбцу стандартное отклонение, 
# допустимый порог отклонения в одну сторону
# (стандартное отклонение, умноженное на 3), среднее значение и значения на расстоянии трех сигм от среднего
def value_range(column):
    std = np.std(column)
    deviation = std * 3
    mean = column.mean()
    min_value = mean - deviation
    max_value = mean + deviation
    return std, deviation, mean, min_value, max_value
```

```python
#проверим работу функции
value_range(df['total_area'])
```

```python
#напишем функцию для фильтрации датафрейма
def df_filter(df, column): #передаем необходимый столбец
    values = value_range(df[column]) #применяем к нему написанную ранее функцию value_range для расчета порогов
    column_min = values[3] #min_value из функции value_range
    column_max = values[4] #max_value из функции value_range
    # фильтруем по следующим условиям: значение больше минимального и меньше максимального, 
    # либо пропущенные значения (для сохранения пропусков в столбце ceiling_height)
    df_filtered = df[(df[column] >= column_min) & (df[column] <= column_max) | (df[column].isna())]
    return df_filtered # возвращает отфильтрованный датафрейм
```

```python
# применим функцию фильтрации к необходимым столбцам
df_filtered = df_filter(df, 'total_area')

df_filtered = df_filter(df_filtered, 'last_price')

df_filtered = df_filter(df_filtered, 'rooms')

df_filtered = df_filter(df_filtered, 'ceiling_height')
```

```python
#сохраним необходимые столбцы с отфильтрованными значениями в отдельный датафрейм
df_main_parameters_filtered = df_filtered[['total_area', 'last_price', 'rooms', 'ceiling_height']]
```

Мы видим сохранившиеся проблемы в данных:

### Анализ десяти населённых пунктов с наибольшим числом объявлений

Выберем 10 населённых пунктов с наибольшим числом объявлений. Посчитаем среднюю цену квадратного метра в этих населённых пунктах и выделим населённые пункты с самой высокой и низкой стоимостью жилья

Проанализируем стобец содержащий названия населенных пунктов:

```python
#добавим столбец в отфильтрованный датафрейм
df_main_parameters_filtered['locality_name'] = df['locality_name']

#изучим распределение
df_main_parameters_filtered['locality_name'].value_counts().head(40)
```

Мы видим различные написания одних и тех же населенных пунктов. Избавимся от указания на тип населенного пункта для приведения названий к единому формату:

```python
# составим список типов населенных пунктов, исходя из того, что они начинаются со строчной буквы
locality_type = [] #создаем пустой список для последующего добавления в него значений
#создадим специальный датафрейм без NaN
df_for_cycle = df_main_parameters_filtered.dropna(subset = ['locality_name']) 
#сбросим индексы для того, чтобы цикл мог пройти по всем значениям без ошибок
df_for_cycle = df_for_cycle.reset_index(drop=True) 
for x in range(len(df_for_cycle['locality_name'])): #напишем цикл для всех значений locality_name
    #зададим переменную words, разбивающую каждое название населенного пункта на отдельные слова
    words = df_for_cycle['locality_name'][x].split()
    #добавим внутренний цикл, проходящий по всем словам внутри каждого названия
    for i in range(len(words)):
        first_letter = words[i][0] #содадим переменную first_letter, содержащую первую букву каждого слова
        if first_letter.islower() == True: #если первая буква строчная
            if words[i] not in locality_type: #если слова с первой строчной буквой нет в списке
                locality_type.append(words[i]) #добавляем это слово в список
            else:
                pass
        else:
            pass
```
